```python
## 'PER'와 'PBR' 컬럼 생성
df_sfs_rn['PER'] = (df_sfs_rn['[공통]PER(최고)(IFRS)'] \
      + df_sfs_rn['[공통]PER(최저)(IFRS)'])/2

df_sfs_rn['PBR'] = (df_sfs_rn['[공통]PBR(최고)(IFRS)'] \
      + df_sfs_rn['[공통]PBR(최저)(IFRS)'])/2
        
```

```python
## '경상연구개발비', ' 임차료 및 리스료', '광고 및 판매촉진비' 컬럼들 삭제
df_sfs_rn = df_sfs_rn.drop(['[공통]   경상연구개발비(IFRS)' \
                            , '[공통]   임차료 및 리스료(IFRS)' \
                            , '[공통]   광고 및 판매촉진비(IFRS)'], axis=1)
```

```python
## 'PBR(최고)(IFRS)', 'PBR(최저)(IFRS)', 'PER(최고)(IFRS)', 'PER(최저)(IFRS)' 컬럼들 삭제
df_sfs_rn = df_sfs_rn.drop(['[공통]PBR(최고)(IFRS)' \
                            , '[공통]PBR(최저)(IFRS)' \
                            , '[공통]PER(최고)(IFRS)' \
                            , '[공통]PER(최저)(IFRS)'], axis=1)
```
