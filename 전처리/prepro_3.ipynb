```python
# 기존 컬럼 활용 새 파생 변수 생성 

df['성장비용'] = df['경상연구개발비'] + df['광고 및 판매촉진비'] + df['임차료 및 리스료']
df['PBR'] = (df['PBR(최고)'] + df['PBR(최저)']) / 2
df['PER'] = (df['PER(최고)'] + df['PER(최저)']) / 2
```

```python
# 파생 변수 생성에 활용한 컬럼들과 불필요한 컬럼 제거

df.drop(['경상연구개발비', '광고 및 판매촉진비', '임차료 및 리스료', 'PBR(최고)', 'PBR(최저)', 'PER(최고)', 'PER(최저)', 'CAPEX', '외국인_주식분포비율',
         '새유형자산합', '감가총계', '유형자산증가액'], axis=1, inplace=True)
```

```python
for i in df.index:
    if pd.isna(df.loc[i, '자산']) or df.loc[i, '자산'] == '':
        df.drop(i, axis=0, inplace=True)
```

```python
counts = df['회사명'].value_counts()
mask = df['회사명'].isin(counts[counts != 9].index)
df.drop(df[mask].index, inplace=True)
```
